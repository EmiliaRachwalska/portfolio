### Problem 1: Lagrange interpolation

The Lagrange interpolation polynomial of degree at most $n$ that passes through the set of points $(x_0, x_1), (x_1, y_1), \dots, (x_n, y_n)$ is the polynomial defined as
$$P(x) = \sum_{k=0}^{n} y_k L_k(x),$$
where
$$L_k(x) = \prod_{\substack{i=0 \\ i \neq k}}^{n} \frac{x - x_i}{x_k - x_i}.$$

#### Part 1 - Implement into code

Write a function `lagrange` that evaluates the Lagrange interpolation polynomial at a given point.

```python
def lagrange(x_points, y_points, x):
    p=0
    for k in range(len(x_points)):
        L=1
        for i in range(len(x_points)):
            if k!=i:
                L*=(x-x_points[i])/(x_points[k]-x_points[i])
        p += y_points[k]*L
    return p
  
```

```python
x_points = [0, 1, 2, 3, 4, 5]
y_points = [0.0, 0.8, 0.9, 0.1, -0.8, -1.0]
print(lagrange(x_points, y_points, 4.5)) 
```

#### Part 2 - Visualisation

Consider the points $(0, 0), (1, 0.8), (2, 0.9), (3, 0.1), (4, -0.8), (5, -1)$. Plot the interpolation polynomial that goes through these points. Overlay the original interpolation points on the graph to visually verify the interpolation. Plot the resulting polynomial over a the range of $x$-values in $[-1, 6]$. 

```python
import matplotlib.pyplot as plt

x_points = [0, 1, 2, 3, 4, 5]
y_points = [0.0, 0.8, 0.9, 0.1, -0.8, -1.0]
xses = [0.01*i for i in range(-100,600,1)]
p = [lagrange(x_points, y_points, i) for i in xses]

plt.plot(figsize=(10,8))
plt.plot(xses, p, label = 'Lagrange Polnomial')
plt.scatter(x_points, y_points, color="orange", label = 'Interpolation Points')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```

#### Part 3 - Least squares polynomial

In addition to constructing and evaluating the Lagrange interpolation polynomial, compute a least squares polynomial of degree 3 that best fits the given set of points.

Add to your previous plot the graph of the least squares polynomial of degree 3 for these points.

```python
import numpy as np

x_points = [0, 1, 2, 3, 4, 5]
y_points = [0.0, 0.8, 0.9, 0.1, -0.8, -1.0]

p3 = np.polyfit(x_points, y_points, 3)
ls_poly_y = [p3[0]*x**3 + p3[1]*x**2 + p3[2]*x + p3[3] for x in xses]

plt.plot(figsize=(10,8))
plt.plot(xses, p, label = 'Lagrange Polnomial')
plt.plot(xses, ls_poly_y, label = 'Least Squares Polnomial')
plt.scatter(x_points, y_points, color="orange", label = 'Interpolation Points')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```
